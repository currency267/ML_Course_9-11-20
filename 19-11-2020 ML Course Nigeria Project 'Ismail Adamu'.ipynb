### Preprocessing data

```python
# checking missing values
data.isnull().sum()
```

```python
dup = data[data.duplicated()]
dup.shape

#print('There are 240 duplicates. But I think it is wise to keep these "extra" as I am assuming that the quality ratings for the same/similar wine were given by different wine tasters.')
```

```python
# separate "quality" as label and the rest as features
y = data['quality']
X = data.drop('quality', axis=1)
print(X.shape, y.shape)
```

```python
# since there's imbalance in label group 'quality'>5 as good wine
y1 = (y>5).astype(int)
y1.head()
```

```python
hs = y1.plot.hist(color='blue')
hs.set_title('Quality  distribution', fontsize=14)
hs.set_xlabel('aggregated label')
plt.show()
```

## Training and Evaluating using Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
#Model
from sklearn.tree import DecisionTreeClassifier
```

```python
#split data into training and test 
X_train, X_test, y_train, y_test = train_test_split(X,y1, test_size=0.20, random_state=43)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train,y_train)
```

```python
print('Accuracy of train: ', clf.score(X_train,y_train))
print('Accuracy of test: ', clf.score(X_test,y_test))
```

```python
pred = clf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
#accuracy score
print('Accuracy score {}'.format(accuracy_score(y_test,pred)))
```

```python
#confusion matrix
print('Confusion matrix {}'.format(confusion_matrix(y_test,pred)))
```

#### Decision Tree has a total of 89 classification error with 72% Accuracy

## Training and Evaluating using Logistic Rgression

```python
#Model using Logistic Regression
#Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
LR_clf = LogisticRegression(solver='liblinear',random_state= 43, max_iter=100)
```

```python
#Train and evaluate Logistic regression using cross val score

cv_scores = cross_val_score(LR_clf, X_train, y_train, cv=10, scoring='accuracy')
print('The cv scores from the iterations are {}'.format(cv_scores))
print('The mean cv score is {}'.format(cv_scores.mean()))

```

```python
#perform predictions
LR_clf.fit(X_train,y_train)
LR_pred = LR_clf.predict(X_test)
```

```python
print(classification_report(y_test,LR_pred))
```

```python
#Accuracy and confusion matrix
print("Accuracy of Logistic {}".format(accuracy_score(y_test,LR_pred)))
print("Confusion matrix of Logistic model {}".format(confusion_matrix(y_test,LR_pred)))
```

#### Logistic Regression has a total of 79 classification error with 75% Accuracy

## Training and Evaluating using Random Forest Classifier

```python
#Using Random Forest 
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_clf = RandomForestClassifier(random_state=7)
RF_clf
```

```python
# Compute k-fold cross validation on training dataset and see mean accuracy score
cv_scores = cross_val_score(RF_clf,X_train, y_train, cv=10, scoring='accuracy')
print('The accuracy scores for the iterations are {}'.format(cv_scores))
print('The mean accuracy score is {}'.format(cv_scores.mean()))
```

```python
#perform predictions 
RF_clf.fit(X_train,y_train)
RF_pred = RF_clf.predict(X_test)
```

```python
print(classification_report(y_test,RF_pred))
```

```python
#Checking accuracy and confusion matrix
print('Accuracy score of Random Forest Classifier {}'.format(accuracy_score(y_test,RF_pred)))
print('Cofusion matrix of Rabdom forest {}',format(confusion_matrix(y_test,RF_pred)))
```

#### Random Forest has a total of 60 classification errors and 81% Accuracy score.
